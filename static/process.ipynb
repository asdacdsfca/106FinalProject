```python
import pandas as pd
```

```python
df = pd.read_json('data.json')

# Display the first few rows of the DataFrame
```

```python
df
```

```python
import statsmodels.api as sm

# Add a constant term for the intercept
X = sm.add_constant(df['EDUC'])
y = df['HOURWAGE']

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary results
print(model.summary())
```

```python
# Assuming your model variable is named model
intercept_statsmodels = model.params[0]  # Intercept (beta_0)
coefficient_statsmodels = model.params[1]  # Coefficient for EDUC (beta_1)

print("Intercept (statsmodels):", intercept_statsmodels)
print("Coefficient for EDUC (statsmodels):", coefficient_statsmodels)


# Calculate predicted HOURWAGE for each EDUC value
df['PREDICTED_HOURWAGE_with_educ'] = intercept_statsmodels + coefficient_statsmodels * df['EDUC']

# Display the DataFrame with predicted values
df
```

```python
import statsmodels.api as sm

# Add a constant term for the intercept
X = sm.add_constant(df[['EDUC', 'AGE']])
y = df['HOURWAGE']

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary results
print(model.summary())
```

```python
# Assuming your model variable is named model
intercept_statsmodels = model.params[0]  # Intercept (beta_0)
coefficient_statsmodels = model.params[1]  # Coefficient for EDUC (beta_1)
coefficient_statsmodels2 = model.params[2]

print("Intercept (statsmodels):", intercept_statsmodels)
print("Coefficient for EDUC (statsmodels):", coefficient_statsmodels)


# Calculate predicted HOURWAGE for each EDUC value
df['PREDICTED_HOURWAGE_with_2'] = intercept_statsmodels + coefficient_statsmodels * df['EDUC'] + coefficient_statsmodels2 *df['AGE']

# Display the DataFrame with predicted values
df
```

```python
# Assuming df has columns named 'AGE' and 'EDUC'
unique_rows = df.drop_duplicates(subset=['AGE', 'EDUC'])

# Display the resulting DataFrame with unique AGE and EDUC combination
unique_rows
```
